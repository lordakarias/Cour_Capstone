```python
import requests
import pandas as pd
import numpy as np
import json
```

```python
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
df = pd.read_html(website_url)
df
df = df[0]
df = df[df.Borough != 'Not assigned']
df
```

```python

df
```

```python
df2 = df.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)
df2
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])
df2
```

<h1>Steps </h1>

<ol>
<li>Send a get request to the url</li>
<li>Read html with pd</li>
<li>Pull the first array of the content to get the table</li>
<li>Cleaned the 'Not assigned' Boroughs</li>
<li>Joinied rows that have the same Postcode</li>
<li>When the Neighborhood has a 'Not assigned' value, gave it Borough value.</li>
</ol>
